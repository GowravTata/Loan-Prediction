```python
import os
os.chdir('F:\Data\loan prediction')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data = pd.read_csv('train.csv')
data  =data.drop(['Loan_ID'],axis=1)
```

```python
data.head()
```

```python
total  = float(len(data))
ax = sb.countplot(x='Education',data=data)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}'.format(height/total),
            ha="center") 
plt.show()
```
