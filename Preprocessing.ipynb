```python
import os
os.chdir('F:\Data\loan prediction')
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.head()
```

```python
def nulldata(data):
    a = (data.isnull().sum()/len(data)*100).reset_index().sort_values(by=0,ascending=False) 
    a = pd.DataFrame(a)
    a['Null_Values'] = (a[0]*len(data))/100
    a['Null_Values']= a['Null_Values'].astype(int)    
    a = a.rename(columns={'index':'Feature',0:'Percent'})
    return a
```

```python
nulldata(train)
```

```python
train = train.drop(['Loan_ID'],axis=1)
```

# As we notice there are not many null values in the dataset , this has to be imputed by mean or mode , as much compirsing categorical data

```python
nulldata(train)
```

```python
def fillna(data):
    for i in data.columns:
        if data[i].dtype=='object':                  
            data[i] = data[i].fillna(data[i].mode()[0])
        else:
             data[i] = data[i].fillna(data[i].mean())            
```

```python
fillna(train)
```

```python
nulldata(train)
```

```python
nulldata(test)
```

```python
fillna(test)
```

```python
nulldata(test)
```

# Label Encoding

```python
def labelencode(data):
    for i in data.columns:
        if data[i].dtype=='object':
            data[i]=LabelEncoder().fit_transform((data[i].values))
```

```python
labelencode(train)
```

```python
labelencode(test)
```
